```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import torch
import numpy as np
import networkx as nx
import dgl
from cycler import cycler
from umap import UMAP
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
from graph_env.environ import GraphEnv
from graph_env.agent import DQN, train, eval, noise_decay
```

```python
plt.rcdefaults()
plt.rcParams['axes.facecolor'] = 'lightgrey'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = 'w'
plt.rcParams['axes.prop_cycle'] = cycler(color='bgrcmyk')
plt.rcParams['font.family'] = 'arial'
```

```python
### Configuration ###

### Environment/Graph ###
n_components = 25
num_communities = 10
n_iter_no_change = 10
max_iter = 100
# min_change = 0.003
min_change = 0.005
# min_change = 0.002
# min_change = 0.0025

### Agent ###
hidden_dim = 400
eta = 0.0001
mem_size = 2048
num_eval_episodes = 10
save_path = './model_cora_4'

### Training ###
burn_in = 512
train_steps = 9
update_freq = 3
batch_size = 16
gamma = 0.95
tau = 0.05
exploration_rate = 0.999
exploration_decay = 1e-5
num_episodes = 2_000
```

```python
graph = dgl.data.CoraGraphDataset()[0]
full_features = csr_matrix(graph.ndata['feat'])
graph.ndata['feat'] = torch.tensor(
    TruncatedSVD(n_components=n_components).fit_transform(full_features)
).to(graph.ndata['feat'])
```

```python
env = GraphEnv(graph, num_communities=num_communities,
               n_iter_no_change=n_iter_no_change, max_iter=max_iter, min_change=min_change)
adapter = env.get_adapter(features=graph.ndata['feat'])
```

```python
agent = DQN(
    adapter,
    hidden_dim=hidden_dim,
    eta=eta,
    mem_size=mem_size,
)
```

---
### Training

```python
preds, eval_history = eval(env, agent, num_episodes=num_eval_episodes, verbose=False)
np.mean([env.compute_modularity(p) for p in preds])
```

```python
train_history = train(
    env=env,
    agent=agent,
    burn_in=burn_in,
    train_steps=train_steps,
    update_freq=update_freq,
    batch_size=batch_size,
    gamma=gamma,
    tau=tau,
    exploration_rate=exploration_rate,
    exploration_decay=exploration_decay,
    num_episodes=num_episodes,
    verbose=True,
    save_path=save_path,
)
```
